# Gather the data and then clean up

```python
#this was a fool hardy task
df = pd.read_csv('dow_jones.csv')
```

```python
def prep_DOW(df):
    df.date = pd.to_datetime(df['date'], dayfirst=True, infer_datetime_format=True)
    df = df.set_index('date').sort_index()
    df = df.replace('\$',' ', regex=True)
    df['next_weeks_close'] = df.next_weeks_close.astype('float64')
    df['next_weeks_open'] = df.next_weeks_open.astype('float64')
    df['open'] = df.open.astype('float64')
    df['high'] = df.high.astype('float64')
    df['low'] = df.low.astype('float64')
    df['close'] = df.close.astype('float64')
    df['quarter'] = df.quarter.astype('int')

    return df
```

```python
m = 58
mp = stumpy.stump(df['percent_change_price'], m)
```

```python
motif_idx = np.argsort(mp[:, 0])[0]
print(f"The motif is located at index {motif_idx}")
```

```python
nearest_neighbor_idx = mp[motif_idx, 1]
print(f"The nearest neighbor is located at index {nearest_neighbor_idx}")
```

```python
fig, axs = plt.subplots(2, sharex=True, gridspec_kw={'hspace': 0})
plt.suptitle('Motif (Pattern) Discovery', fontsize='30')
axs[0].plot(df['percent_change_price'].values)
axs[0].set_ylabel('percent_change_price', fontsize='20')
rect = Rectangle((643, 0), m, 40, facecolor='lightgrey')
axs[0].add_patch(rect)
rect = Rectangle((800, 0), m, 40, facecolor='lightgrey')
axs[0].add_patch(rect)
axs[1].set_xlabel('date', fontsize ='20')
axs[1].set_ylabel('Matrix Profile', fontsize='20')
axs[1].axvline(x=643, linestyle="dashed")
axs[1].axvline(x=800, linestyle="dashed")
axs[1].plot(mp[:, 0])
```

```python
group_single = df.groupby('stock').agg({'close':['mean','min','max']})
group_single.columns = ['close_mean','close_min','close_max']
group_single = group_single.reset_index()
print(group_single.sort_values(by='close_mean'))
```

```python
stock_change = df.groupby('stock').agg({'percent_change_price':['mean','min','max']})
stock_change.columns = ['percent_change_mean',
                        'percent_change_min',
                        'percent_change_max']
stock_change = stock_change.reset_index()
print(stock_change.sort_values(by='percent_change_mean'))
```

```python
stock_volume = df.groupby('stock').agg({'volume':['mean','min','max']})
stock_volume.columns = ['volume_mean',
                        'volume_min',
                        'volume_max']
stock_volume = stock_volume.reset_index()
print(stock_volume.sort_values(by='volume_mean'))
```
